# **Filling the null values in Age column with Mean**

```python
mean=dataset["Age"].mean()
mean
```

```python
dataset["Age"]=dataset["Age"].fillna(mean)
```

```python
dataset["Age"].isnull().sum()           #null values are replaced with mean
```

# **Filling the Null values in Embarked with mode**

```python
dataset["Embarked"].isnull().sum()
```

```python
Em_mode=dataset["Embarked"].mode()
```

```python
dataset["Embarked"]=dataset["Embarked"].fillna(Em_mode[0])
```

```python
dataset["Embarked"].isnull().sum()       #null values are replaced with mode
```

# **5.Handling the outliers**


```python
sns.boxplot(dataset["Age"])
```

```python
Age_q1 = dataset.Age.quantile(0.25)
Age_q3 = dataset.Age.quantile(0.75)
print(Age_q1)
print(Age_q3)
```

```python
IQR_Age=Age_q3-Age_q1
IQR_Age
```

```python
upperlimit_Age=Age_q3+1.5*IQR_Age
upperlimit_Age
```

```python
lower_limit_Age = Age_q1-1.5*IQR_Age
lower_limit_Age
```

```python
median_Age=dataset["Age"].median()
median_Age
```

```python
dataset["Age"]=np.where(dataset["Age"]>upperlimit_Age,median_Age,dataset["Age"])
```

```python
(dataset["Age"]>54.5).sum()
```

```python
sns.boxplot(dataset["Age"])
```

```python

dataset["Age"]=np.where(dataset["Age"]<lower_limit_Age,median_Age,dataset["Age"])
```

```python
sns.boxplot(dataset["Age"])
```

```python
sns.boxplot(dataset["Fare"])
```

```python
Fare_q1 = dataset.Fare.quantile(0.25)
Fare_q3 = dataset.Fare.quantile(0.75)
print(Fare_q1)
print(Fare_q3)
```

```python
IQR_Fare=Fare_q3-Fare_q1
IQR_Fare
```

```python
upperlimit_Fare=Fare_q3+1.5*IQR_Fare
upperlimit_Fare
```
